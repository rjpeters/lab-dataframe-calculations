## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

```python
orders['CustomerID'].nunique()
# I need to compare 4339 customers.
```

```python
temporder = orders.groupby(['CustomerID']).agg({'amount_spent':sum})
# Storing this in a new dataframe to keep working with it
temporder.head()
```

```python
# How to select customers whose aggregated `amount_spent` is in a given quantile range?

#define the quantiles:
vipcutoff = temporder.quantile(.95)

prefcutoff = temporder.quantile(.75)

vip = temporder[temporder['amount_spent']>vipcutoff[0]]
pref = temporder[(temporder['amount_spent'] > prefcutoff[0]) & (temporder['amount_spent'] < vipcutoff[0])]
```

```python
# How to label selected customers as "VIP" or "Preferred"

# My goto way of doing this is generating a list, which I will then use to define a new column
customertype = []

# I also need to convert the VIP and Preferred customer IDs into lists
viplist = list(vip.index)
preflist = list(pref.index)


for i in orders.index:
    if orders['CustomerID'][i] in viplist:
        customertype.append('VIP')
    elif orders['CustomerID'][i] in preflist:
        customertype.append('Preferred')
    else: 
        customertype.append('Regular')
        
orders['CustomerType'] = customertype
```

## Q2: How to identify which country has the most VIP Customers?

```python

temp = pd.concat([orders['Country'], orders['CustomerType']], axis = 1)

vipnumbers = temp['Country'][temp['CustomerType'] == 'VIP'].value_counts()
display(vipnumbers)
# This gives the absolute number of VIP customers, but how does this relate to the total number of customers from each country?
```

```python

totals = temp['Country'].value_counts()
display(totals)

# I can see right away that there are a number of countries without any VIP customers, so I'll only need the total counts for countries that do have VIPs.
```

I'll use a right join to put 'vipnumbers' and 'totals' into a single dataframe. This first requires some cleaning of the dataframes...

(At first I transformed 'totals' to only include countries also present in 'vipnumbers', but then found out that this makes it hard to reuse the 'totals' series for the next assignment.)

```python
# Convert to dataframe (found out the hard way after fiddling around with series that this is the easier way)
totalsframe = pd.DataFrame(data=totals)
vipframe = pd.DataFrame(data=vipnumbers)

# Reset the index to get the 'Country' column back into the dataframe
totalsframe = totalsframe.reset_index()
vipframe = vipframe.reset_index()

# Rename the columns to match the country
totalsframe = totalsframe.rename(columns = {"index": "Country", "Country": "Total"})
vipframe = vipframe.rename(columns = {"index": "Country", "Country": "VIPs"})


# Merge the two (with a right join)

newframe = pd.merge(  left = totalsframe,
                        right = vipframe,
                        how = 'right', 
                        left_on = "Country", 
                        right_on= "Country")

# Calculate the ratio of VIPs to total:
newframe['Ratio'] = newframe['VIPs'] / newframe['Total']

# Sort the result by highest ratio
newframe.sort_values(['Ratio'], ascending = False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# Again I'll do the rough-and-ready method for getting absolute numbers first: 

temp = pd.concat([orders['Country'], orders['CustomerType']], axis = 1)

combonumbers = temp['Country'][(temp['CustomerType'] == 'VIP') | (temp['CustomerType'] == 'Preferred')].value_counts()
display(combonumbers)
```

Then I want to get the ratios again. I still have the dataframe 'totalsframe', so only have to go through part of the motions.


```python
# First order of business: get the dataframe again

# Initialize a dataframe
comboframe = pd.DataFrame(data=combonumbers)

# Reset the index to get the 'Country' column back into the dataframe
comboframe = comboframe.reset_index()

# Rename the columns to match the country
comboframe = comboframe.rename(columns = {"index": "Country", "Country": "VIP and Preferred"})

# Now left join total frame onto combo frame

newframe2 = pd.merge(
                        left = totalsframe,
                        right = comboframe,
                        how = 'right', 
                        left_on = "Country", 
                        right_on= "Country")
```

```python
# Calculate the ratio of VIPs to total:
newframe2['Ratio'] = newframe2['VIP and Preferred'] / newframe2['Total']

# Sort the result by highest ratio
newframe2.sort_values(['Ratio'], ascending = False)
```
